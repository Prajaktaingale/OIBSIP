# Import dataset

```python
df = pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv', sep=';')
```

# Get unique values (class or label)in y variables

```python
df['quality'].value_counts()
```

# *define y(dependent or label or taeget variable)and x (independent or features or attributr variable)

```python
y = df['quality']
```

```python
y.shape
```

```python
y
```

```python
x = df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
```

```python
x.shape
```

```python
x
```

# Get x variables standardized

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

# Get train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, stratify= y, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Get Model Train

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train, y_train)
```

# Get Model Prediction

```python
y_pred = svc.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# Get Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
